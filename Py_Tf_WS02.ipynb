Working with tensors and visualizing

```python
import tensorflow as tf
import numpy as np
```

```python
tensor = tf.zeros((200, 300, 3), dtype = tf.int32)
tensor2 = tf.random.uniform((200, 300, 3),minval = 0, maxval = 255, dtype = tf.int32)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(tensor)
```

```python
plt.imshow(tensor2)
```

```python
rad = np.linspace(0, 2 * np.pi, 100)
```

```python
plt.plot(rad, np.sin(rad), label = "Sin")
plt.plot(rad, np.cos(rad), c ='r', marker = 'D', label = "Cos")
plt.legend()
plt.show()
```
